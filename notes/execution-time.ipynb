# Timing Python code

````python
def f(n):
    a = 0
    for i in range(100_000 * n):
        a += i
    return a
````

````python
%time f(30)
````

````python
# also demos use of varargs!
import time
def gettime(f, *args):
    "Return time in seconds; %time, %timeit don't work in functions so we write our own"
    t0 = time.time()
    f(*args)
    t1 = time.time()
    return t1 - t0
````

````python
t = gettime(f, 30) # call f(30)
print(f"{t:.3f}s")
````

## Plotting time vs work

Let's compute and plot how long something takes for a variety of input sizes with:

```python
times = [gettime(f, i) for i in range(1,n,step)]
```

The visualization will give us an idea of whether it's linear or quadratic etc.

````python
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('svg') # png sucks
def showtime(f, n, step=1):
    times = [gettime(f, i) for i in range(1,n,step)]
    plt.figure(figsize=(4,3))
    plt.plot(range(1,n,step), times, '.', color='#fdae61', alpha=.9, markersize=15, markeredgewidth=.5, markeredgecolor='black')
    plt.xlabel("Problem size")
    plt.ylabel("Time in seconds")
    plt.show()
````

````python
showtime(f,10)
````

````python
def g(n):
    a = 0
    for i in range(10*n):
        zeroes = [0 for j in range(i)]
        a += len(zeroes)
    return a

showtime(g,400,step=10)
````

# Misc

````python
def log2(n):
    if n <= 1:
        return 0
    count = 0
    while n > 0:
        n = n//2
        count += 1
    return count-1
print( log2(2), log2(32) )
````
